```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import squarify
print("Successful")
```

```python
BigBasket = pd.read_csv("BigBasketRefined.csv", header = None)
BigBasket.head()
```

```python
plt.rcParams['figure.figsize'] = (18,7)
color = plt.cm.copper(np.linspace(0,1,40))
BigBasket[0].value_counts().head(40).plot.bar(color = color)
plt.title('Graphical representation of Most Popular Items')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud

plt.rcParams['figure.figsize'] = (15, 15)
wordcloud = WordCloud(background_color = 'white', width = 1200,  height = 1200, max_words = 121, stopwords = ('Description', 'NAN','columns')).generate(str(BigBasket))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Most Popular Items',fontsize = 20)
plt.show()
```

```python
BigBasket_Count = BigBasket[0].value_counts().head(50).to_frame()
BigBasket.keys()
```

```python
plt.rcParams['figure.figsize'] = (20, 30)
color = plt.cm.cool(np.linspace(0, 1, 60))
squarify.plot(sizes = BigBasket_Count.values, label = BigBasket_Count.index, alpha=.8, color = color)
plt.title('Tree Map for Popular Items')
plt.axis('off')
plt.show()
```
